```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
import jax
import jax.numpy as jnp

plt.rcParams.update({
    # Axes
    "axes.spines.left": True, # display axis spines
    "axes.spines.bottom": True,
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.grid": True,
    "axes.axisbelow": True,
    # Fonts
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "font.size": 10,
    "text.usetex": True,
    # Grid
    "grid.linestyle": "-",
    "grid.linewidth": 0.2,
    "grid.color": "cbcbcb"
})
```

```python
A = (pd.read_csv(
    "https://raw.githubusercontent.com/dynamicslab/databook_python/master/DATA/hald_ingredients.csv", 
    delimiter=",", 
    header=None)
).values
b = (pd.read_csv(
    "https://raw.githubusercontent.com/dynamicslab/databook_python/master/DATA/hald_heat.csv", 
    delimiter=",", 
    header=None)
).values
```

```python
# OLS
x = np.linalg.inv(A.T @ A) @ A.T @ b
```

```python
plt.figure(figsize=(7, 3))
plt.plot(b, color="blue", label="Observed")
plt.plot(A @ x, "--", color="red", label="Predicted")
plt.legend();
```

```python
# OLS with SVD
U, S, VT = np.linalg.svd(A, full_matrices=0)
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b
```

```python
plt.figure(figsize=(7, 3))
plt.plot(b, color="blue", label="Observed")
plt.plot(A @ x, "--", color="red", label="Predicted")
plt.legend();
```
