```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

### Figure 3.5

```python
data = []
for n in range(2, 10000):
    data.append(np.random.normal(loc=2., scale=10., size=n))

sample_mean = [np.mean(sample) for sample in data]

plt.plot(np.arange(2, 10000), np.cumsum(sample_mean) / np.arange(2, 10000))
plt.show()
```

```python
data = []
for n in range(2, 10000):
    data.append(np.random.pareto(1.13, n))

sample_mean = [np.mean(sample) for sample in data]

plt.figure(figsize=(10, 5))
plt.plot(np.arange(2, 10000), np.cumsum(sample_mean) / np.arange(2, 10000))
plt.axhline(y=1.13 / (1.13 - 1), color='green', label='true mean')
plt.axhline(y=np.mean(sample_mean), color='red', label='sample mean')
plt.legend()
plt.show()
```
